
# Task 3

1)**Create a histogram or bar char to vizualize the distribution of price range amoung the Restaurants**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/Dataset .csv')
plt.hist(df['Price range'])
plt.xlabel('Price Range')
plt.ylabel('No. of Restaurant')
plt.title('Distribution of Price Range')
plt.show()


```

### **2)Calculate the percentage of restaurants in each price range category**

```python
count=df['Price range'].value_counts()
percentage=(count/len(df))*100
print(percentage)
```
